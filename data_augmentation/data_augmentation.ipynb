```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
print(tf.__version__)

import os
os.chdir("./")

image_path = './cat.jpg'
```

```python
image = tf.io.read_file(image_path)
image = tf.io.decode_image(image)
print(image.shape)

plt.imshow(image)
```

```python
image2 = tf.image.adjust_brightness(image, 0.4)
plt.imshow(image2)
```

```python
image3 = tf.image.flip_left_right(image)
plt.imshow(image3)
```
